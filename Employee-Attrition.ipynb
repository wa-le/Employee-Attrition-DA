Employee attrition problem\
Employees are very important in the organization. Organization's performance is highly based on the quality of the employees\
An Organization can face various challenges due employee attrition, and they are:

1. Expensive in terms of both money and time to train new employees.
2. Loss of experienced employees
3. Impact in productivity
4. Impact profit

Business Questions:
1. What factors are contributing more to employee attrition?
2. What type of measures should the company take in order to retain their employees?

```

%sql
With AttritionCount (select Attrition, count(Attrition)as AttrCount from emp_attrition
group by Attrition)
select * from AttritionCount where Attrition = "Yes";
```

```
%sql

with final as (
  select age, sum(employeecount) as sumofempcount,
  case 
    when age between 18 and 24 then '18-24'
    when age between 25 and 31 then '25-31'
    when age between 32 and 38 then '32-38'
    when age between 39 and 45 then '39-45'
    when age between 46 and 52 then '46-52'
    else '52+'
  end as age_group
  from emp_attrition
  where Attrition = 'Yes'
  group by age, age_group
)
select age_group, sum(sumofempcount)
from final
group by age_group
```

```
%sql
select * from emp_attrition limit 1;
select distinct(BusinessTravel) from emp_attrition;

select BusinessTravel, sum(employeecount) from emp_attrition
where Attrition == 'Yes'
group by BusinessTravel;
```

```
%sql
select * from emp_attrition limit 1;

-- check distribution of daily rate
select distinct(DailyRate) from emp_attrition;
select max(distinct(DailyRate)) from emp_attrition;

-- spliting daily rate (1-500 as LOW | 501-1000 as MID | 1001-1500 as HIGH)

with cte (select DailyRate, sum(employeecount) as sumofempcount,
  case 
    when DailyRate between 0 and 300 then 'LOW'
    when DailyRate between 501 and 1000 then 'MID'
    else 'HIGH'
  end as DailyRateGroup
  from emp_attrition
  where Attrition = 'Yes'
  group by DailyRateGroup, DailyRate)
select DailyRateGroup, sum(sumofempcount)
from cte
group by DailyRateGroup;
```

```
%sql
select * from emp_attrition limit 1;
select distinct(gender) from emp_attrition;

with Gender (select gender, count(gender) as total_gender from emp_attrition
where Attrition == 'Yes'
group by gender
order by total_gender)
select gender, total_gender, round((total_gender / 237) * 100, 1) as Percentage
from Gender;
```

```

%sql
-- Education
--    1 - No College
--    2 - College
--    3 - Bachelor Degree
--    4 - Master Degree
--    5 - PhD

With cte (select education, count(employeenumber) as total_employees from emp_attrition
where Attrition == 'Yes'
group by education
order by total_employees)
select
  case
    when education == 1 then "No College Degree"
    when education == 2 then "College Degree"
    when education == 3 then "Bachelor Degree"
    when education == 4 then "Master Degree"
    else "PhD"
  end as EducationLevel,
total_employees from cte;
```

```

%sql
-- (1-low, 2-medium, 3-high, 4-highly satisfied)
select EnvironmentSatisfaction, count(employeenumber) as total_employees from emp_attrition
where attrition='Yes'
group by 1
order by 1 desc;
```

```

%sql
-- (1-low, 2-medium, 3-high, 4-highly satisfied)
select distinct(jobsatisfaction) from emp_attrition;

with cte (select jobsatisfaction, count(employeenumber) as total_employees from emp_attrition
where attrition='Yes'
group by 1
order by 1 desc)
select
  case
    when jobsatisfaction == 1 then "low"
    when jobsatisfaction == 2 then "medium"
    when jobsatisfaction == 3 then "high"
    else "highly satisfied"
  end as EducationLevel,
total_employees from cte;
```

```
%sql
select MaritalStatus, count(employeenumber) as total_employees from emp_attrition
where attrition='Yes'
group by 1
order by 1 desc;
```
